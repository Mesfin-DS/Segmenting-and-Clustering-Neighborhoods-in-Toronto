```python
import pandas as pd
import numpy as np
```

```python
# load the geo data provided by coursera
df_geo = pd.read_csv('https://cocl.us/Geospatial_data')
df_geo.head()
```

```python
# change the column name
df_geo = df_geo.rename(columns = {'Postal Code': 'Postalcode'})
df_geo.head()
```

```python
# load the clean data done in part 1
df_clean_final = pd.read_csv('https://cocl.us/Geospatial_data').drop(['Unnamed: 0'], axis = 1)
df_clean_final.head()
```

```python
# join df_clean_final and df_geo by Postalcode
df_combined = pd.merge(df_clean_final, df_geo, on = 'Postalcode')
df_combined.head()
```

```python
df_combined.shape
```

```python
#Save data as'Capstone_part2.csv'
df_combined.to_csv('capstone_part2.csv', index=False)
print('Saved Successfully!')
```
